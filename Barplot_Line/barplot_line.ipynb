```python
import matplotlib.pyplot as plt

# Assicuriamoci che i dati siano ordinati
df = df.sort_values(by='ppda', ascending=False)

# Crea il grafico
fig = plt.figure(figsize=(16, 19), facecolor='blanchedalmond')
# Crea il grafico delle barre sull'asse principale
ax = fig.add_axes([0, 0.2, 1, 0.7])  # [left, bottom, width, height]
ax.set_facecolor('blanchedalmond')
ax.grid(True, which='both', axis='both', linestyle='--', linewidth=0.5, alpha=0.5, color = "grey")

# Colori per le barre: un colore per tutte, e un colore speciale per il Milan
#colors = [(139/255, 0/255, 0/255, 0.3) if team != 'Bologna' else 'darkred' for team in df['Squad']]


# Crea il grafico a barre
bars = ax.bar(df["team"], df['ppda'], color= "darkred", alpha= 0.5, zorder = 2)

# Aggiunge i valori sopra ogni barra
for bar, value in zip(bars, df['ppda']):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.08,
             f'{value:.1f}', ha='center', va='bottom', fontsize=26, color='#003153', fontproperties = font_prop)

# Crea un secondo asse per il grafico a linee
ax2 = ax.twinx()
ax2.plot(range(0,20), df['Def. event height (m)'], color='#003153', linewidth=2, marker='o', markersize=10)


# Migliora le etichette degli assi
ax.set_ylabel("PPDA", fontproperties=font_prop_bold, fontsize=30,color='#003153')
ax2.set_ylabel("Altezza media interventi difensivi (m)", fontproperties=font_prop_bold, 
               fontsize=30, color='#003153')
#plt.xticks(rotation=90, fontproperties=font_prop, fontsize=18, color='#003153')
#plt.yticks(fontproperties=font_prop, fontsize=20, color='#003153')
ax.tick_params(axis='both', which='major', labelsize=26)
ax.set_xticklabels(df['team'], rotation=90, fontproperties=font_prop, fontsize=18, color='#003153')
ax.set_yticklabels(ax.get_yticks(), fontproperties=font_prop, fontsize=26, color='#003153')

ax2.tick_params(axis='both', which='major', labelsize=26)
ax2.set_yticklabels(ax2.get_yticks(), fontproperties=font_prop, fontsize=26, color='#003153')

ax.set_xticks([])

# Crea un asse secondario per i loghi
ax_loghi = fig.add_axes([0, 0.07, 1, 0.1])  # Posiziona subito sotto l'asse principale
ax_loghi.axis('off')  # Rimuove bordi e assi
ax_loghi.set_xlim(ax.get_xlim())  # Allinea l'asse x dei loghi con quello principale

# Aggiunge i loghi all'asse dei loghi
for i, logo_path in enumerate(df['Logo']):
    logo_img = plt.imread(logo_path)  # Carica l'immagine del logo
    image_box = OffsetImage(logo_img, zoom=0.38)  # Regola la dimensione del logo
    ab = AnnotationBbox(image_box, (i, 1), frameon=False, box_alignment=(0.5, 0.5))
    ax_loghi.add_artist(ab)


ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines["bottom"].set_color('dimgrey')
ax.spines["left"].set_color('dimgrey')

ax2.spines['top'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax2.spines["bottom"].set_color('dimgrey')
ax2.spines["right"].set_color('dimgrey')


ax.text(-2.3,19,"Quanti passaggi si concedono agli avversari? E fino a che punto?", fontsize=50, color='darkred', fontproperties = font_prop_bold)
ax.text(-2.3,17.8,"Il Como gioca da grande squadra ma ha gli stessi punti del Verona e forse non è una caso che le due che\nintervengono difensivamente più alte giocano a 3 dietro", fontsize=30, color='darkorange', fontproperties = font_prop_bold)
plt.text(16,0.2,"Dati Soccerment  |  @AlfoMarino0975", fontsize=24, color='grey', fontproperties=font_prop)
#plt.text(-1.5, 0.5, "Cambi di gioco:", fontsize = 22, color = "goldenrod", fontproperties = font_med)
#plt.text(0.75, 0.5, "passaggi che percorrono piu di 35 metri in ampiezza.", fontsize = 20, color = "gray", fontproperties = font_prop)

#plt.tight_layout()

#plt.savefig("/Users/alfonsomarino/Desktop/ppda_barplot.png", bbox_inches='tight', dpi=300)

plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

def calculate_weighted_fei_fii(df):
    """
    Calcola gli indici FEI (Finalization Efficiency Index) e FII (Final Involvement Index)
    per ogni giocatore in un DataFrame che include metriche offensive per 90 minuti.

    Parametri:
    ----------
    df : pd.DataFrame
        DataFrame contenente le colonne:
        'G/Sh', 'G-xG', 'SoT', 'Gls', 'Touches', 'xA', 'KP', 'PrgP', 'Att Pen'

    Ritorna:
    --------
    df_result : pd.DataFrame
        DataFrame con due nuove colonne: 'FEI' e 'FII'
    """
    # Definisci i pesi
    fei_weights = {"G-xG": 0.35, "SoT": 0.30, "G/Sh": 0.20, "Gls": 0.15}
    fii_weights = {"xA": 0.30, "PrgP": 0.25, "Att Pen": 0.20, "KP": 0.15, "Touches": 0.10}

    # Copia il DataFrame per non modificarlo in-place
    df_res = df.copy()

    # Rimuove eventuali stringhe, sostituisce valori non numerici con NaN e poi con 0
    for col in list(fei_weights.keys()) + list(fii_weights.keys()):
        df_res[col] = pd.to_numeric(df_res[col], errors="coerce").fillna(0)

    # Z-score
    scaler = StandardScaler()
    for col in fei_weights.keys():
        df_res[col + "_z"] = scaler.fit_transform(df_res[[col]])
    for col in fii_weights.keys():
        df_res[col + "_z"] = scaler.fit_transform(df_res[[col]])

    # Calcolo indice FEI
    fei = sum(df_res[col + "_z"] * w for col, w in fei_weights.items())
    df_res["FEI"] = fei / sum(fei_weights.values())

    # Calcolo indice FII
    fii = sum(df_res[col + "_z"] * w for col, w in fii_weights.items())
    df_res["FII"] = fii / sum(fii_weights.values())

    return df_res
```
